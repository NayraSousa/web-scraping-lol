```python
import pandas as pd
import requests
import os
from urllib.parse import unquote, urlparse
```

```python
df = pd.read_csv('voicelines(6).csv')
```

```python
if not os.path.exists('downloaded_audios'):
    os.makedirs('downloaded_audios')
```

```python
def extrair_nome_arquivo(url):
    caminho = urlparse(url).path
    nome_arquivo = unquote(caminho.split('/')[-3])
    return nome_arquivo
```

```python
def baixar_link(url, nome_arquivo):
    try:
        response = requests.get(url)
        response.raise_for_status()
        with open(nome_arquivo, 'wb') as arquivo:  # Salva o conteúdo em um arquivo
            arquivo.write(response.content)
        print(f"Saved file: {nome_arquivo}")
    except requests.exceptions.RequestException as e:
        print(f"Error when downloading {url}: {e}")
```

```python
for indice, linha in df.iterrows():
    url = linha['audio_link']
    nome_base = extrair_nome_arquivo(url)
    nome_arquivo = f"downloaded_audios/file_{indice}_{nome_base}"
    baixar_link(url, nome_arquivo)
```
